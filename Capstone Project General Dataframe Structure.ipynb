Players need to be 19 years old and one year removed from highschool. However, there are cases of players playing in overseas basketball leagues after highschool and entering the NBA, or repeating a year in highschool.**

```python
df1.loc[(df1['College'].isnull()) & (df1['Country'] == 'USA') & (df1['Year'] < 2006), 'College'] = 'Highschool'
```
