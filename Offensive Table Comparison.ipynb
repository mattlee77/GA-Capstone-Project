```python
offlr = joblib.load('Offense_LR.pkl')
offrg = joblib.load('Offense_Ridge.pkl')
offls = joblib.load('Offense_LS.pkl')
```

```python
def getpredictdfoff(draftdf,featuredf,year,model1,model2,model3):
    tomerge = draftdf[['Year','Pick','Team','Name','College']].copy()
    draftyear = tomerge.loc[tomerge['Year'] == year].copy()
    names = list(draftyear['Name'])
    collegedf = featuredf.loc[featuredf['Name'].isin(names)].copy()
    draftcomp = draftyear.merge(collegedf, on = 'Name')
    x = draftcomp[[u'PPG', u'Games Played 4', u'FGM', u'FG%', u'FTMSR', u'FT%C', u'3PMSR',
       u'3P%', u'ORBSR', u'SPGSR', u'BPG', u'TOVSR', u'PF', u'TopSchool',
       u'Age at Draft SR', u'Weight at Draft', u'Height Inches C']]
    predictlr = model1.predict(x)
    predictrg = model2.predict(x)
    predictls = model3.predict(x)
    draftcomp['PredictionsLR'] = predictlr
    draftcomp['PredictionsRG'] = predictrg
    draftcomp['PredictionsLS'] = predictls
    draftcompfinal = draftcomp[['Year_x','Pick','Name','Team_x','College','TotalOff','PredictionsLR','PredictionsRG',
                                'PredictionsLS']].copy()
    draftcompfinal = draftcompfinal.sort_values(by = 'PredictionsRG', ascending = False)
    draftcompfinal = draftcompfinal.reset_index(drop = True)
    draftcompfinal = draftcompfinal.rename_axis('Model Pick')
    draftcompfinal = draftcompfinal.reset_index()
    draftcompfinal['Model Pick'] = draftcompfinal['Model Pick'] + 1
    return draftcompfinal
```

**We will get predictions for the 2017 NBA Draft**

```python
earlyentry = pd.read_csv('/Users/mattlee/GA-Capstone-Project/earlyentrydf1.csv')
```

```python
def getpredictearlyoff(df,model1,model2,model3):
    earlydf = df
    x = df[[u'PPG', u'Games Played 4', u'FGM', u'FG%', u'FTMSR', u'FT%C', u'3PMSR',
       u'3P%', u'ORBSR', u'SPGSR', u'BPG', u'TOVSR', u'PF', u'TopSchool',
       u'Age at Draft SR', u'Weight at Draft', u'Height Inches C']]
    predictlr = model1.predict(x)
    predictrg = model2.predict(x)
    predictls = model3.predict(x)
    earlydf['PredictionsLR'] = predictlr
    earlydf['PredictionsRG'] = predictrg
    earlydf['PredictionsLS'] = predictls
    earlydf = earlydf.sort_values(by = 'PredictionsRG', ascending = False)
    earlydf = earlydf.reset_index(drop = True)
    earlydf = earlydf.rename_axis('Model Pick')
    earlydf = earlydf.reset_index()
    earlydf['Model Pick'] = earlydf['Model Pick'] + 1
    return earlydf[['Model Pick','Name','Team','PredictionsLR','PredictionsRG','PredictionsLS']]
```

```python
d2017 = getpredictearlyoff(earlyentry,offlr,offrg,offls)
d2017
```
