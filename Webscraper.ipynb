```python
main_table = content.find('tbody')
```

```python
data = {}
player_row = main_table.find_all('tr')
for i, player in enumerate(player_row):
    data[i] = {}
    year = player.find_all('td')
    if len(year[0]) == 0 or year[0] is None:
        data[i]['Year'] = 'MISSING'
    else:
        data[i]['Year'] = year[0].text
    if len(year[1]) == 0 or year[1] is None:
        data[i]['Draft Position'] = 'MISSING'
    else:
        data[i]['Draft Position'] = year[1].text
    data[i]['NBA Team'] = year[2]['data-order']
    data[i]['Name'] = year[3].find('a')['title'].split(' draftexpress')[0]
    if len(year[4]['data-order']) == 0 or year[4]['data-order'] is None:
        data[i]['Country'] = 'MISSING'
    else:
        data[i]['Country'] = year[4]['data-order']
    if len(year[5]) == 0 or year[5] is None:
        data[i]['Position'] = 'MISSING'
    else:
        data[i]['Position'] = year[5].text
    if len(year[6]) == 0 or year[6] is None:
        data[i]['Age at Draft'] = 'MISSING'
    else:
        data[i]['Age at Draft'] = year[6].text
    if len(year[7]) == 0 or year[7] is None:
        data[i]['Height at Draft'] = 'MISSING'
    else:
        data[i]['Height at Draft'] = year[7].text
    if len(year[8]) == 0 or year[8] is None:
        data[i]['Weight at Draft'] = 'MISSING'
    else:
        data[i]['Weight at Draft'] = year[8].text
    if len(year[9]) == 0 or year[9] is None:
        data[i]['Wingspan'] = 'MISSING'
    else:
        data[i]['Wingspan'] = year[9].text
    
df = pd.DataFrame().from_dict(data, orient = 'index')
df.to_csv('draftexpress2016.csv', index = False)



```

```python
with open('earlyentryid.html', 'r') as f:
    content = BeautifulSoup(f.read())
```

```python
data = {}
table = content.find('tbody')
player_row = table.find_all('tr')
for i, player in enumerate(player_row):
    data[i] = {}
    row = player.find_all('td')
    data[i]['ID'] = row[1].find('a')['title'].split(' ')[0].split('-')[2]
    data[i]['Name'] = row[1].find('a').text.split('\n\t\t\t\t\t\t\t')[1].split('\n')[0]

df = pd.DataFrame().from_dict(data, orient = 'index')
df.to_csv('earlyentryid.csv', index = False)
```
