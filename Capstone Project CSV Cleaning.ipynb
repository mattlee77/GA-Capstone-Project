```python
import re

asdf = 'br1991draft.'
re.findall('\d+', asdf)[0]
```

```python
def brcleaner(csv):
    dataframes = []
    for c in csv:
        if c == 'br1991draft.csv':
            df = pd.read_csv('~/project-capstone/Data/%s' % c)
            cols = ['Rank','Pick','Team','Name','College','Seasons','Games','Minutes Played','Points','Total Rebounds',
               'Assists','FG%','3P%','FT%','MPG','PPG','TRPG','APG','Win Share','Win Share/48','Box +/-','VORP']
            df.columns = cols
            df = df.iloc[1:].copy()
            df.drop('Rank', axis = 1, inplace = True)
            df['Year'] = re.findall('\d+', c)[0]
            dataframes.append(df)
        else:
            df1 = pd.read_csv('~/project-capstone/Data/%s' % c, header = None)
            cols1 = ['Rank','Pick','Team','Name','College','Seasons','Games','Minutes Played','Points','Total Rebounds',
               'Assists','FG%','3P%','FT%','MPG','PPG','TRPG','APG','Win Share','Win Share/48','Box +/-','VORP']
            df1.columns = cols1
            df1 = df1.iloc[2:,1:].copy()
            df1['Year'] = re.findall('\d+', c)[0]
            dataframes.append(df1)
    return dataframes

def dfjoiner(listdf):
    df = pd.concat(listdf).reset_index(drop = True)
    cols = ['Year','Pick','Team','Name','College','Seasons','Games','Minutes Played','Points','Total Rebounds',
               'Assists','FG%','3P%','FT%','MPG','PPG','TRPG','APG','Win Share','Win Share/48','Box +/-','VORP']
    df = df[cols].copy()
    return df
```

```python
csvs = []
for rge in range(1991, 2017):
    csvs.append('br%sdraft.csv' % rge)
dfs = brcleaner(csvs)
df = dfjoiner(dfs)
df.to_csv('brdraft.csv', index = False)
df.shape
```

Cleaning/Compiling NCAA Individual Stats CSVs
==

```python
def ncaacleaner(csv):
    dataframes = []
    for c in csv:
        df = pd.read_csv('~/project-capstone/Data/%s' % c)
        df['Year'] = re.findall('\d+', c)[0]
        dataframes.append(df)
    return dataframes

def dfjoiner(listdf):
    df = pd.concat(listdf).reset_index(drop = True)
    return df
```

```python
csv = []
for rge in range(2003, 2017):
    csv.append('realgmncaa%s.csv' % rge)
dfs = ncaacleaner(csv)
df = dfjoiner(dfs)
df.to_csv('realgmncaa.csv', index = False)
df.shape
```
